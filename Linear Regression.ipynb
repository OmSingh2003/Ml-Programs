```python
import pandas as pd #importing datasets
import numpy as np # computation
import matplotlib.pyplot as plt # graph plotting
import seaborn as sns # mathmatical statistic
```

```python
data = pd.read_csv('/Users/omsingh/Downloads/Datasets/Salary_data.csv')
data.head()
```

```python
plt.scatter(data['YearsExperience'],data['Salary']) # plotting linear regression graph with x as year of exp and y as salary
```

```python
#data extraction
x = data[['YearsExperience']]
y = data['Salary']
# assigning x and y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
xtest
```

```python
pred = lr.predict(xtest)
```

```python
lr.predict([[20]])
```

```python
#finding the prediction 
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Assuming pred contains the predicted values and ytest the true values
mse = mean_squared_error(ytest, pred)
mae = mean_absolute_error(ytest, pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
```

```python
#plotting the graph
plt.scatter(xtrain,ytrain,color='red')
plt.plot(xtrain,lr.predict(xtrain),color ='blue')
plt.xlabel('Years Of Exp')
plt.ylabel('Salary')
plt.show()
```
