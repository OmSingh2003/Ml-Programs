### Multi Variable Linear Regression

```python
import pandas as pd #importing datasets
import numpy as np # computation
import matplotlib.pyplot as plt # graph plotting
import seaborn as sns # mathmatical statistics
```

```python
data1 = pd.read_csv('/Users/omsingh/Downloads/Datasets/advertising.csv')
data1.head()
```

```python
plt.scatter(data1['TV'],data1['Sales']) # plotting linear regression graph with x as TV of exp and y as Sales
```

```python
plt.scatter(data1['Newspaper'],data1['Sales']) # plotting linear regression graph with x as Newspaper of exp and y as Sales
```

```python
#extracting input and output features
x1 = data1[['TV','Radio','Newspaper']]
y1 = data1['Sales']
```

```python
print(x1)
print(y1)
```

```python
from sklearn.model_selection import train_test_split
x1train, x1test, y1train, y1test = train_test_split(x1, y1, test_size=0.3, random_state=42)
```

```python
#import the model
from sklearn.linear_model import LinearRegression
lr1 = LinearRegression()
lr1.fit(x1train,y1train)
```

```python
x1test
```

```python
pred1 = lr1.predict(x1test)
```

```python
#finding the prediction 
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Assuming pred contains the predicted values and ytest the true values
mse1 = mean_squared_error(y1test, pred1)
mae1 = mean_absolute_error(y1test, pred1)

print("Mean Squared Error:", mse1)
print("Mean Absolute Error:", mae1)
```

```python
#plotting the graph
plt.scatter(x1train,y1train,color='red')
plt.plot(x1train,lr1.predict(x1train),color ='blue')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```
